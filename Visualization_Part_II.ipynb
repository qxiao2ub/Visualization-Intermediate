# 3 Scatter Plots

3.1 A basic scatter plot


```python
# We are going to create a basic scatter plot
import matplotlib.pyplot as plt
```

```python
city = ['Buffalo','Dallas','Los Angeles','St. Louis']
x = [50,25,5,32]
y = [30,15,5,20]

demand = [5000,9000, 10000, 3000]

plt.title('Widget Demand')

plt.scatter(x , y, s=demand)
```

3.2 Changing colors based on demand

Suppose we want our demand bubbles to be colored differently, based on the magnitude of demand at each location.

```python
# Changing colors and including transparency
colors = []

for i in range(0,len(demand)):
  colors.append(demand[i]/float(max(demand)))

  myx = x[i]
  myy = y[i]
  mycity = city[i]

  #Formatting text labels: Suppose we want our text labels to be centered in the bubbles. This is easy to do by editing the plt.text().
  plt.text(myx, myy, mycity, color = 'red', fontsize = 12, horizontalalignment='center', verticalalignment='center')

plt.scatter(x, y, s=demand, c=colors, alpha=.5)

```

3.3 Adding text labels

It is difficult to tell which buble goes with which city. Let's label our bubbles with the name of each city.

To do this, we will need to loop over all of the cities to add individual text labels.

```python
for i in range(0,len(city)):
  myx = x[i]
  myy = y[i]
  mycity = city[i]
  plt.text(myx, myy, mycity, color = 'red', fontsize = 12)
```

For the above, we cannot run it separately, because the scatter plot should be added up with different functions.

# 4 Bar Charts

Let's return to our weather example and plot Buffalo's monthly high temperature in a bar chart.

Following is a code snippet for a basic bar chart.

```python
import matplotlib.pyplot as plt

y = [31,31,41,52,63,73,79,78,70,58,48,35]
x = list(range(1,13))

months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

plt.bar(x,y, tick_label = months, width=0.8)
plt.xticks(x,months)
```

# 7. Creating Multiple Plots

Suppose we want to create one figure containing multiple plots (i.e., subplots). We weill need to use "axes" within matplotlib to do this function.

For example, let's create two separate line plots (with markers) showing the monthly high and low temperatures in Buffalo.

```python
# Creating subplots

import matplotlib.pyplot as plt

# Create our data
x = list(range(1,13))
y_high = [31,31,41,52,63,73,79,78,70,58,48,35]
y_low = [20,19,27,38,49,59,66,64,57,45,37,26]

# Labels for the months:
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

# Create two subplots sharing the same y axis
fig, (ax1, ax2) = plt.subplots(2, sharey=True)

# Add a title for the entire figure:
fig.suptitle('Buffalo Monthly Weather')

# Plot the highest temperatures in the top subplot:
ax1.plot(x,y_high,'g.-')

# Add a title for the top subplot, and a lable for its y-axis:
ax1.set(title='Highest Temperatures', ylabel='Highest Temperature (F)')

# Remove the tick marks for the x-axis in the top subplot:
ax1.set_xticks([])

# Plot the low temperatures in the bottom subplot:
ax2.plot(x,y_low,'ro-')

# Add a label for the x-axis in the bottom subplot, and provide a lable for the bottom y-axis:
ax2.set(xlabel='Month', ylabel='Lowest Temperature (F)')

# Set the tickmarks in the bottom subplot to show each x value:
ax2.set_xticks(x)

# Replace the numeric x-axis values with our months:
ax2.set_xticklabels(months)
```
